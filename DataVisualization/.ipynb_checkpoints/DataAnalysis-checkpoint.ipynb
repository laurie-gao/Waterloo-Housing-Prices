```python
import seaborn as sns
sns.set()
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
housing = pd.read_csv('waterlooHousingSummary.csv')
```

```python
housing.head()
```

```python
housing.drop(['Unnamed: 5','Unnamed: 6'], axis=1, inplace=True)
```

```python
housing['price'] = housing['price'].str.replace('$', '')
housing['price'] = housing['price'].str.replace(' ', '')
```

```python
housing['price'] = housing['price'].astype(int)
```

```python
housing[housing['house type']=='Detached']
```

```python
housing.drop(83, inplace=True)
```

```python
print(housing['price'].mean())
```

```python
print('number of condos: ' + str(len(housing[housing['house type']=='Condo'])))
print('number of houses: ' + str(len(housing[housing['house type']=='House'])))
print('number of townhouses: ' + str(len(housing[housing['house type']=='Townhouse'])))
```

```python
housing.rename(columns={'# of beds': '# of bedrooms'}, inplace=True)
```

```python
jointplot = sns.jointplot(x='# of bedrooms', y='price', data=housing)
jointplot.savefig("jointplot.png")
```

```python
sns_boxplot = sns.boxplot(x='house type', y='# of baths', data=housing)
plt.savefig("boxplot.png")
```

```python
sns.heatmap(housing.corr(), cmap='coolwarm')
plt.savefig('heatmap.png')
```

```python
graph = sns.pairplot(housing)
plt.savefig("pairplot.png")
```
